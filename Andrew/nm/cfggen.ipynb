```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import numpy as np
```

```python
# can vary :
epochs = 800
replicas = 15
sets = 50

df = pd.DataFrame()
params = np.array([[1,2,.5,.5]])
batch = 0
line = 0
for hyperparams in params:
    for setno in range(sets):
        for repititions in range(4):
            cfg = pd.DataFrame(np.reshape(hyperparams,(1,4)), columns = ['alpha', 'gamma', 'rho', 'sigma'])
            cfg['set'] = setno
            cfg['epochs'] = epochs
            cfg['replicas'] = replicas
            cfg['line'] = line
            cfg['batch'] = batch
            line = line + 1
            df = df.append(cfg)
        batch = batch + 1
```
